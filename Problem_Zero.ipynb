# Problem Zero

A number is a perfect square, or a square number, if it is the square of a positive integer.
For example, $25$ is a square number because $5 \times 5 = 25$
; it is also an odd square.

The first 5 square numbers are: $1, 4, 9, 16, 25$
, and the sum of the odd squares is $1 + 9 + 25 = 35$
.

Among the first 491 thousand square numbers, what is the sum of all the odd squares?

```python
# Hard-coded solution
def sum_odd_squares(N):
    '''
    This function sums all the odd squares among the first N square numbers.
    '''
    # to keep track on the sum
    accumulated_odd_square_sum = 0
    for n in range(N):
        n += 1 # starts with 1
        n_sq = n**2

        # if odd square, adds to the accumulated sum
        if (n_sq % 2) == 1:
            accumulated_odd_square_sum += n_sq
    
    return accumulated_odd_square_sum

sum_odd_squares(491000)
```
