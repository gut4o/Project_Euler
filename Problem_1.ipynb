# Problem 1 - Multiples of 3 or 5

If we list all the natural numbers below $10$ that are multiples of $3$ or $5$, we get $3 , 5, 6$ and $9$. The sum of these multiples is $23$.

Find the sum of all the multiples of $3$ or $5$ below $1000$.


```python
# Solution
def sum_multiples_3_and_5(N):
    '''
    This function returns the sum of all multiples of 3 or 5 below N.
    '''
    # initialization 
    acc_sum_3 = 0
    acc_sum_5 = 0

    # loop to find multiples of 3 or 5
    for n in range(N):

        # checks if multiple of 3
        if (n % 3 == 0):
            acc_sum_3 += n
        # checks if multiple of 5
        elif (n % 5 == 0):
            acc_sum_5 += n
    
    return acc_sum_3 + acc_sum_5

sum_multiples_3_and_5(1000)
```
